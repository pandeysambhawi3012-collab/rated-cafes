```python
#Ques 3 Given two dataset , now preprocess / clean and merge the data set together 
import numpy as np
import pandas as pd 

patients = pd.read_csv("patient_details.csv")
records = pd.read_csv("medical_records.csv")

records['sugar_level'].fillna(records['sugar_level'].mean(),inplace=True)
print("Medical record Column :\n")


print(records.to_string())
patients['name'].fillna("Chirag",inplace=True)
patients.loc[patients['age'] < 0, 'age'] = np.nan

patients['age'].fillna(patients['age'].mean(),inplace=True)


print("\nPatient Details Columns:\n")
print(patients.to_string())



records['sugar_level'] = pd.to_numeric(
    records['sugar_level'], errors='coerce'
)

Q1 = records['sugar_level'].quantile(0.25)
Q3 = records['sugar_level'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = records[
    (records['sugar_level'] < lower_bound) |
    (records['sugar_level'] > upper_bound)
]
print(outliers.to_string())


print("Outliers:")
print(outliers.to_string())


merged_df = pd.merge(patients,records,on="patient_id", how="outer")

print("\nMerged Data:")
print(merged_df)
```
