```python
#ques 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Student': ['sambhawi','Anshika','Vivaan','Rashi','Aayusha','Arnav','Suman','Tanya','Arya','Chirag','Jiya'],
    'Marks': [89, 67, 78, np.nan, 67, 98, 61, 96, 75, 45, 89],
    'Age': [20, 19, 17, 19, np.nan, 19, 19, 18, 17, np.nan, 20],
    'Subject': ['Phy','Chem','Maths','AI-DS','Python','PHE','Electrical',np.nan,'Maths','Cse', np.nan,]
}

df = pd.DataFrame(data)

df['Age'].fillna(df['Age'].median(), inplace=True)
df['Marks'].fillna(df['Marks'].median(), inplace=True)
df['Subject'].fillna("AI-ML", inplace=True)


print("Cleaned DataFrame:\n", df)


print("\nShape of DataFrame:", df.shape)
print("\nMissing values count:\n", df.isnull().sum())

plt.figure(figsize=(12,8))
plt.bar(df['Subject'], df['Marks'], color='pink')
plt.xlabel("Subject", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.title("Marks Distribution by Subject", fontsize=14)
plt.show()
```
